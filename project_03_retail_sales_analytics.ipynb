```python
import pandas as pd
sales_data = {
    "Store": [
        "Manchester", "London", "Leeds", "Manchester", "London",
        "Leeds", "Manchester", "London", "Leeds", "Manchester"
    ],
    "Region": [
        "North", "South", "North", "North", "South",
        "North", "North", "South", "North", "North"
    ],
    "Product": [
        "Laptop", "Phone", "Headphones", "Phone", "Laptop",
        "Mouse", "Keyboard", "Phone", "Laptop", "Mouse"
    ],
    "Quantity": [
        2, 3, 5, 1, 2,
        4, 3, 2, 1, 6
    ],
    "Price": [
        800, 500, 80, 500, 800,
        25, 50, 500, 800, 25
    ]
}
```

```python
sales = pd.DataFrame(sales_data)

sales
```

```python
sales["Revenue"] = sales["Quantity"] * sales["Price"]

sales
```

```python
store_revenue = sales.groupby("Store")["Revenue"].sum()

store_revenue
```

```python
store_revenue = sales.groupby("Store")["Revenue"].sum()

store_revenue
```

```python
sales.groupby("Product")["Revenue"].sum()
```

```python
sales.groupby("Region")["Revenue"].sum()
```

```python
store_revenue = sales.groupby("Store")["Revenue"].sum()

store_revenue = store_revenue.sort_values(ascending=False)

store_revenue
```

```python
product_revenue = sales.groupby("Product")["Revenue"].sum()

product_revenue.sort_values(ascending=False)
```

```python
product_revenue.sort_values(ascending=True)
```

```python
sales["Revenue"].mean()
```

```python
sales.groupby("Store")["Store"].count()
```

```python
sales["Revenue"].max()
```

```python
sales["Revenue"].min()
```

```python
total_revenue = sales["Revenue"].sum()

average_revenue = sales["Revenue"].mean()

highest_sale = sales["Revenue"].max()

lowest_sale = sales["Revenue"].min()
```

```python
print("Total Revenue:", total_revenue)
print("Average Revenue:", average_revenue)
print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
```

```python
summary = pd.DataFrame({
    "Business Question": [
        "Total Revenue",
        "Average Revenue",
        "Highest Sale",
        "Lowest Sale"
    ],
    "Answer": [
        total_revenue,
        average_revenue,
        highest_sale,
        lowest_sale
    ]
})

summary
```

```python
highest_store = sales.groupby("Store")["Revenue"].sum().idxmax()

highest_product = sales.groupby("Product")["Revenue"].sum().idxmax()

highest_region = sales.groupby("Region")["Revenue"].sum().idxmax()

busiest_store = sales.groupby("Store")["Store"].count().idxmax()
```

```python
summary = pd.DataFrame({
    "Business Question": [
        "Total Revenue",
        "Average Revenue",
        "Highest Sale",
        "Lowest Sale",
        "Highest Revenue Store",
        "Highest Revenue Product",
        "Highest Revenue Region",
        "Store With Most Sales"
    ],
    "Answer": [
        total_revenue,
        average_revenue,
        highest_sale,
        lowest_sale,
        highest_store,
        highest_product,
        highest_region,
        busiest_store
    ]
})

summary
```

```python
summary.to_csv("output/retail_sales_summary.csv", index=False)
```
